## All dataset

```python
stareNormalIDs = np.array([ 77,  81,  82, 162, 163, 235, 236, 239, 240, 255])
driveNormalIDs = np.array([25,26,32])

def VSI_disease_class_all_dataset(label):
    '''
    
    return 0 for Normal, 1 for Diseased
    '''
    
    
    dSet = label[0]
    
    if dSet == "d":
        #Drive
        num = int(label[1:])
        print(num)
        if num not in driveNormalIDs:
            return 0
        else:
            return 1
    
    elif dSet == "h":
        #HRF
        if label[-1] == "h":
            return 0
        else:
            return 1
        
    elif (dSet == "c" or dSet == "k"):
        #CHASE
        return 0
    elif (dSet == "a" or dSet == "v"):
        #STARE
        
        num = int(label[1:])
        if num in stareNormalIDs:
            return 0
        else:
            return 1
        
```

```python
import matplotlib.pyplot as plt
mat = np.load("../all_images.npy",allow_pickle=True).item()

X = {}

keys = list(mat.keys())

for key in keys:
    
    X[key] = [VSI_disease_class_all_dataset(key)]

    
diagnosis_codes = {0:"Normal",1:"Diseased"}
diagnosis_abbrevs = {0:"Normal",1:"D"}
    
data = {}
data['image_diagnoses'] = X
data['diagnosis_codes'] = diagnosis_codes
data['diagnosis_abbrevs'] = diagnosis_abbrevs

    
np.save("image_diagnoses_all.npy",data)    
```
